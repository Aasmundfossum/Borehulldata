```python
import pandas as pd
import numpy as np 

import folium
from pyproj import Proj, transform
```

```python
df = pd.read_excel("Testoversikt.xlsx", sheet_name="Alle tester",usecols='D,W,AH,AI')
```

```python
df = df.replace('-','0')
df = df.loc[df['Dybde [m]\n'].astype(float) >=245]
df = df.dropna()
df = df.reset_index(drop=True)
#print(df)
```

```python
utm_projection = Proj(proj='utm', zone=32, ellps='WGS84')

# Function to convert UTM to lat-lon
def utm_to_latlon(easting, northing):
    lon, lat = utm_projection(easting, northing, inverse=True)
    return pd.Series({'latitude': lat, 'longitude': lon})

# Apply the conversion function to the DataFrame
df[['latitude', 'longitude']] = df.apply(lambda row: utm_to_latlon(row['ØV (UTM 32)\n'], row['NS (UTM 32)\n']), axis=1)

# Print the result
print(df)
```

```python
map_osm = folium.Map(location=[65.5, 18], zoom_start=4.5)
df.apply(lambda row:folium.CircleMarker(location=[row["latitude"], row["longitude"]],radius=3).add_to(map_osm), axis=1)
map_osm.save("Brønner_dypere_enn_245_m.html")
map_osm
```

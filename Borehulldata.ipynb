```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import streamlit as st

filename = 'GrunnvannsBorehull.xlsx'
```

```python
energibronn = pd.read_excel(filename, sheet_name = 'EnergiBr√∏nn', usecols='G')
```

```python
print(type(energibronn))    

energibronn = energibronn.to_string()

print(energibronn)
```

```python
startaar = 1995
sluttaar = 2023
aarstall = np.arange(startaar,sluttaar+1)
antall_per_aar = np.zeros(len(aarstall))

for aar in range(startaar,sluttaar+1):
    antall_per_aar[aar-startaar] = energibronn.count(str(aar)+'-')

print(antall_per_aar)
```

```python
fig = plt.figure()
plt.bar(aarstall, antall_per_aar)
plt.xlabel("√Örstall")
plt.ylabel("Antall br√∏nner boret")
plt.title("Antall br√∏nner boret i Norge per √•r")
plt.show()
st.pyplot(fig)
```

```python
steder = pd.read_excel(filename, sheet_name = 'Br√∏nnPark', usecols='C')
steder = steder.to_string()

steder = steder.split()
print(steder)
```

```python
steder.sort()
print(steder)
```

```python
unike_steder = ['xx']
for i in range(0,len(steder)):
    if steder[i] != unike_steder[-1]:
        unike_steder.append(steder[i])
        unike_steder = unike_steder
print(unike_steder)

unike_steder = unike_steder[1:]
```

```python
from string import digits

def remove_tall(list):
    list = [''.join(x for x in i if x.isalpha()) for i in list]
 
    return list
 
unike_steder = remove_tall(unike_steder)

while("NaN" in unike_steder):
    unike_steder.remove("NaN")

while("" in unike_steder):
    unike_steder.remove("")   

print(unike_steder)
print(len(unike_steder))
```

```python
antall_per_sted = np.zeros(len(unike_steder))
for i in range(0,len(unike_steder)):
    antall_per_sted[i] = steder.count(unike_steder[i])

print(antall_per_sted)
print(len(antall_per_sted))
```

```python
fig_steder = plt.figure()
plt.bar(unike_steder, antall_per_sted)
plt.xlabel("Steder")
plt.ylabel("Antall br√∏nner boret")
plt.title("Antall br√∏nner boret ulike steder i Norge")
plt.show()
st.pyplot(fig_steder)
```

```python
import numpy as np
import matplotlib.pyplot as plt
import csv


filename ='GrunnvannsBorehull_kikketpaa.csv'

oppdragstaker = [] 
with open(filename) as csvfile:
    csvReader = csv.reader(csvfile, delimiter=';')
    for row in csvReader:
        oppdragstaker.append(row[5])

oppdragstaker.sort()

unike_oppdrag = ['oppdragstaker']
for i in range(0,len(oppdragstaker)):
    if oppdragstaker[i] != unike_oppdrag[-1]:
        unike_oppdrag.append(oppdragstaker[i])
#        unike_bedrifter = unike_bedrifter
unike_oppdrag = unike_oppdrag[1:]

print(unike_oppdrag)
print(len(unike_oppdrag))

antall_per_oppdrag = np.zeros(len(unike_oppdrag))
for i in range(0,len(unike_oppdrag)):
    antall_per_oppdrag[i] = oppdragstaker.count(unike_oppdrag[i])

fig_bedrifter = plt.figure()
plt.bar(unike_oppdrag, antall_per_oppdrag)
plt.xlabel("Oppdragstaker")
#plt.xticks(rotation = -90)
plt.ylabel("Antall br√∏nner boret")
plt.title("Antall br√∏nner boret av ulike oppdragstakere i Norge")
plt.show()
#st.pyplot(fig_bedrifter)
```

```python
storste_oppdrag = []
storste_antall_per_oppdrag = []
for i in range(0,len(unike_oppdrag)):
    if antall_per_oppdrag[i]>=1000:
        storste_oppdrag.append(unike_oppdrag[i])
        storste_antall_per_oppdrag.append(antall_per_oppdrag[i])

fig_storste_bedrifter = plt.figure()
plt.bar(storste_oppdrag, storste_antall_per_oppdrag)
plt.xlabel("Oppdragstaker")
plt.xticks(rotation = -90)
plt.ylabel("Antall br√∏nner boret")
plt.title("Antall br√∏nner boret av ulike oppdragstakere i Norge")
plt.show()
```
